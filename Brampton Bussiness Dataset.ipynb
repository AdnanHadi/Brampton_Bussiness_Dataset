# Brampton Business Dataset

The Brampton Business Directory is an employer census of all brick and mortar businesses operating within the city and features over 8,900 records.  The directory contains information such as company name, business description, address, website and North American Industry Classification Code, the year the business was established in Brampton and the year the business was established as a business._

*http://geohub.brampton.ca/datasets/brampton-business-directory*

## ***Analysis***

Using this dataset we were able to determine:

1) top 10 employers in Brampton: (accounts for 15% of total employment in Brampton) 22,998 employees (of 150,413 employees in Brampton)

2) top 1% of employers in Brampton (97 companies), generate 35.5% of all employment in Brampton
